# Measure Tempo

DrumScript estimates the tempo (BPM) of any audio file using a **spectral onset envelope** approach.

Under the hood, `ds.detect_tempo()` works by:
1. Computing an **onset strength envelope** — a curve that peaks wherever a drum hit occurs.
2. Building a **tempogram** — a matrix that measures how periodic (i.e. "beat-like") the onset envelope is at every possible BPM.
3. Summing the tempogram energy across time, then picking the BPM with the strongest peak — but only within a plausible musical range (60–240 BPM) to avoid artifacts.

If the audio is shorter than 1 second, tempo estimation is unreliable, so DrumScript defaults to 120 BPM.

*(Note: We are using a copyright-free synthetic track for this demonstration.)*
